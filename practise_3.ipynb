visualizing using matplotlib

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
from matplotlib import axis


plt.plot(10, 5, '*') 
```

```python
x = np.array([2,4,6,8,10,12,14,16])

y = x/2   
print(y)

#Total points -- (2,1),(4,2), (6,3), (8,4),(10,5),(12,6),(14,7),(16,8)

plt.figure(figsize=(10,6))
plt.scatter(x, y, c='red')
plt.ylabel('Y-data')
plt.xlabel('X-data')
plt.show()
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show()
```

```python
names = ['jhon', 'Sanjay', 'Ellen','Sandra','Nancy','Shubham','Philip']
scores = [10,15,20,30,40,50,100]

plt.figure(figsize=(20, 10))

plt.subplot(231)        
plt.bar(names, scores)
plt.ylabel('column2')
plt.xlabel('column1')

plt.subplot(232)
plt.scatter(names, scores)

plt.subplot(233)
plt.plot(names, scores)

plt.subplot(234)
plt.bar(names, scores)

plt.subplot(235)
plt.scatter(names, scores)


plt.suptitle('Categorical Plotting')       #you can give titles,xlabels and ylabels to each of the plots as well
plt.show()
```

```python
import seaborn as sns
import pandas as pd
```

```python
ourdata=pd.read_csv(r"C:\Users\DELL\Downloads\pokemon.csv")
```

```python
ourdata.head()
```

```python
# top5 pokemon power representation

top_pokemon = ourdata.sort_values(by="sp_defense", ascending=False).head(5)

plt.figure(figsize=(10,6))

plt.bar(top_pokemon["sp_attack"], top_pokemon["sp_defense"])

plt.xlabel("attack")
plt.ylabel("defense")
plt.title("Top 5 Most Powerful Pokemon")

plt.xticks(rotation=90)

plt.show()
```

```python
sns.lmplot(data=ourdata,x = 'sp_attack', y = 'sp_defense')
```

```python
ourdata['is_legendary'].value_counts()
```

```python
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1]) #The calling signature of add_axes is add_axes(rect), where rect is a list [x0, y0, width, height] denoting the lower left point of the new axes in figure coodinates (x0,y0) and its width and height. So the axes is positionned in absolute coordinates on the canvas

x = np.arange(1,10)
a1.plot(x, np.exp(x),'b')

a1.set_title('range of numbers')

#explicitly set x and y limits
plt.ylim(0,10000)
plt.xlim(0,15)

#explicitly set x and y labels
plt.xlabel("x-axis") 
plt.ylabel('y-axis')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(data=ourdata[['sp_attack','sp_defense']])
```

```python
sns.pairplot(ourdata[['sp_defense','sp_attack','speed']],kind='scatter')
plt.show()
```
