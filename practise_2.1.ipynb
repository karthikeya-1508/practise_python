How to fill values 

    -> forwardfill => sequential repeated values
    
    -> backwardfill => sequential repeated values
    
    -> fillna => if we know what to fill
    
    -> replace => replacing the null values wi=th someother value
    
    -> mean => replacing with mean if any extreme values then use median
    
    -> median => filling null value with middle value
    
    -> mode => using relation formula r most repeated one
    
    -> random => In set of values choosing one value
    
    -> imputer methods> sklearn 
    
    -> min -> minimum 
    
    -> max -> maximum value in the set of data

```python
import pandas as pd
data = {
    'A': [1, None, None, 4, 5],
    'B': [None, 2, None, 4, None]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df_ffill = df.ffill()
```

```python
df_ffill
```

```python
df_bfill = df.bfill()
```

```python
df_bfill
```

```python
# Fill null values with a specific value (e.g., 0)
df_fillna_with_spec_num = df.fillna(0)
df_fillna_with_spec_num
```

```python
# Fill null values with the mean of the column
mean_value = df['A'].mean()
print("Mean value of column A:", mean_value)
df_mean_filled = df['A'].fillna(mean_value)
df_mean_filled
```

```python
# Fill null values with the median of the column
median_value = df['A'].median()
print("Median value of column A:", median_value)
df_median_filled = df['A'].fillna(median_value)
df_median_filled
```

```python
# Fill null values with the mode of the column
mode_value = df['A'].mode()[0]
print("Mode value of column A:", mode_value)
df_mode_filled = df['A'].fillna(mode_value)
df_mode_filled
```

```python
# Fill null values in column 'A' with its minimum value
min_value = df['A'].min()
df_min_filled = df['A'].fillna(min_value)
df_min_filled

```

```python
# Fill null values in column 'A' with its maximum value
max_value = df['A'].max()
df_max_filled = df['A'].fillna(max_value)
df_max_filled
```

Scikit-learn provides several imputer methods for handling missing values:

SimpleImputer: Fills missing values with a specified strategy (mean, median, most frequent, constant).

KNNImputer: Fills missing values using the k-nearest neighbors algorithm.

IterativeImputer: Fills missing values using multivariate regression (estimates each feature as a function of others).

MissingIndicator: Creates a boolean indicator for missing values (used for feature engineering).

```python
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.experimental import enable_iterative_imputer

imputer = SimpleImputer(strategy='mean')
df_imputed = imputer.fit_transform(df)
df_imputed
```

```python
knn_imputer = KNNImputer(n_neighbors=2)
df_knn_imputed = knn_imputer.fit_transform(df)
df_knn_imputed
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

iter_imputer = IterativeImputer()

df_iter_imputed = iter_imputer.fit_transform(df)

df_iter_imputed
```
