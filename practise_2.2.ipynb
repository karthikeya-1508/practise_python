group by

```python
import pandas as pd

data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
print(df.groupby('Company')['Sales'].sum())
```

```python
print(df.groupby('Company')['Sales'].mean())
```

```python
print(df.groupby('Company')['Sales'].max())
```

```python
print(df.loc[df.groupby('Company')['Sales'].idxmax()])
```

```python
grouped = df.loc[df.groupby('Company')['Sales'].idxmax()]
```

```python
grouped
```

```python
df = pd.DataFrame(abs(randn(1000,4)),columns='Sales Trasaction Units Gross_Profits'.split())
```

```python
df['Gross_Profits'].hist()
```

```python
df['Sales'].plot(kind = 'hist',bins = 20)
```

```python
df['Sales'].plot(kind = 'line')
```

- 'line' : line plot (default)
    - 'bar' : vertical bar plot
    - 'barh' : horizontal bar plot
    - 'hist' : histogram
    - 'box' : boxplot
    - 'kde' : Kernel Density Estimation plot
    - 'density' : same as 'kde'
    - 'area' : area plot
    - 'pie' : pie plot
    - 'scatter' : scatter plot (DataFrame only)
    - 'hexbin' : hexbin plot (DataFrame only)

```python
df.plot.area()
```

```python
df.plot.bar()
```

```python
df.plot.area(stacked =False)
```

```python
df.plot.hist(bins = 40)
```

```python
df.plot.scatter(x='Sales' , y='Units')
```

```python
df.plot.scatter(x='Sales' , y='Units', c="Trasaction",cmap='coolwarm')
```

```python
df.plot.scatter(x='Sales' , y='Units', s=df["Trasaction"]*100)
```
