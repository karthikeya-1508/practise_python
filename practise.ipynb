```python
str = "abba"
rev_str = str[::-1]

if  str == rev_str:
    print("string is palindrome")
else:
    print("not a palindrome")
```

```python
def check_palindrome(s):
  s = s.replace(" ", "").lower()
  return s == s[::-1]

input_str = input("Enter a string: ")
if check_palindrome(input_str):
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

```python
def split_the_string(s):
     
    a = s.upper()
    #print(a)
    print('Converting the given string into upper case: {}'.format(a))
    b = a.lower()
    #print(b)
    print('Converting the given string into lower case: {}'.format(b))
    c = b.split()
    #print(c)
    print('Converting the given string into split case: {}'.format(c))
    return a,b,c
```
