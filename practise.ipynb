```python
str = "abba"
rev_str = str[::-1]

if  str == rev_str:
    print("string is palindrome")
else:
    print("not a palindrome")
```

```python
def check_palindrome(s):
  s = s.replace(" ", "").lower()
  return s == s[::-1]

input_str = input("Enter a string: ")
if check_palindrome(input_str):
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

```python
def split_the_string(s):
     
    a = s.upper()
    #print(a)
    print('Converting the given string into upper case: {}'.format(a))
    b = a.lower()
    #print(b)
    print('Converting the given string into lower case: {}'.format(b))
    c = b.split()
    #print(c)
    print('Converting the given string into split case: {}'.format(c))
    return a,b,c
```

```python
split_the_string(s)
```

```python
def is_prime(n):
    
    for i in range(2,n):
      if n % i == 0:
        print(f"{n} is not a prime number")
        break
      else:
        print(f"{n} is a prime number")
        break
```

```python
is_prime(67)
```

```python
def simple_calculator(a,b,operation):
   if operation == "add":
     return a + b
   elif operation == "subtract":
     return a - b
   elif operation == "multiply":
     return a * b
   elif operation == "division":
     return a / b
   else:
      return "invalid operation"
```

```python
simple_calculator(61,6,"add")
```

```python
def get_even_or_odd_numbers(nums):
  even_numbers = []
  odd_numbers = []

  for num in nums:
    if num %2 == 0:
      even_numbers.append(num)
    else:
      odd_numbers.append(num)
  return even_numbers, odd_numbers  
```

```python
get_even_or_odd_numbers([2,3,4,5,6,7,8,2,556,765,344,8399])
```
