concatenation

```python
#suppose i need to merge two tables. df1 & df2 are two dataframes.we use concatenation  method to merge two tables. we can use concat() function of pandas library to merge two tables. we can specify the axis parameter to specify the direction of concatenation. if axis=0, then the tables will be merged vertically. if axis=1, then the tables will be merged horizontally.

df1 = pd.DataFrame({
  "A" : [1, 2, 3],
  "B" : ["a", "b", "c"],
  "C" : [10, 20, 30]
})

df2 = pd.DataFrame({
  "A" : [4, 5, 6],
  "B" : ["d", "e", "f"],
  "C" : [40, 50, 60]
})

df = pd.concat([df1, df2], ignore_index=True)
```

merged,joins

```python
df1 = pd.DataFrame({
  "Key" : ['k0', 'k1', 'k2'],
  "A" : ['A0', 'A1', 'A2'],
  "B" : ['B0', 'B1', 'B2']
})
```

```python
df2 = pd.DataFrame({
  "Key" : ['k1', 'k2', 'k3'],
  "C" : ['C0', 'C1', 'C2'],
  "D" : ['D0', 'D1', 'D2']
})
```

```python
df = pd.merge(df1, df2,how = 'inner', left_on='Key',right_on='Key')
```

```python
df
```

```python
df = pd.merge(df1, df2,how = 'left', left_on='Key',right_on='Key')
df
```

```python
df = pd.merge(df1, df2,how = 'right', left_on='Key',right_on='Key')
df
```

```python
df = pd.merge(df1, df2,how = 'outer', left_on='Key',right_on='Key')
df
```
