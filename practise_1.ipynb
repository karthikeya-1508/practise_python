copy & view

```python
a = np.array([1,2,3,4,5])

b = a.view()

b[0] = 10

print(a)

#original array is also changed when we use view() method
```

```python
a = np.array([1,2,3,4,5])

b = a.copy()

b[0] = 10

print(a)

#original array is not changed when we use copy() method 
```

stacking

```python
arr1 = np.array([[1, 2],[3, 4]])
arr1
```

```python
arr2 = np.array([[5,6],[7,8]])
arr2
```

```python
array_hstack = np.hstack((arr1, arr2))
array_hstack
```

```python
array_vstack = np.vstack((arr1, arr2))
array_vstack
```

```python
array_col = np.column_stack((arr1, arr2))
array_col
```

```python
array_col.flatten()
```

```python
array_col.flatten(order = "f")
```
